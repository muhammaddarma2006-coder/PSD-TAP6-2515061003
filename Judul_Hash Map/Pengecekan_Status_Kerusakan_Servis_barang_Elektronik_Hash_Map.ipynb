```python
def main():

    status_servis = {
        "SRV001": "Selesai (Ganti LCD iPhone 13)",
        "SRV002": "Sedang Dikerjakan (Solder IC Power Laptop)",
        "SRV003": "Menunggu Part (Lensa Kamera Sony)",
        "SRV004": "Selesai (Pembersihan Debu PS5)",
    }

    print("=== PROGRAM PENGECEKAN STATUS SERVIS ELEKTRONIK ===")

    kode_cari = input("Masukkan nomor tiket servis yang ingin dicek: ").strip().upper()

    if kode_cari in status_servis:

        print(f"Status untuk nomor tiket {kode_cari} adalah: {status_servis[kode_cari]}")
    else:

        print("Kode barang tidak ditemukan!")

if __name__ == "__main__":
    main()
```
